```python
# Linera Graph

import matplotlib.pyplot as plt
plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
## Graph to plot distance and time
distance=list(range(2,11,2))
time=list(range(1,6))
#distance
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Distance vs Time')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
#v=a t
t=list(range(1,11))
v=[2*i for i in range(1,11)]
#plt.plot(time,distance,'bo')
plt.plot(t,v)
plt.grid()
plt.show()
v
```

```python
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
## NUMBERS VS CUBES



n=list(range(10))
#n
plt.title('NUMBER VS CUBES')
sqn=[i**3 for i in n]
plt.plot(n,sqn)
```

```python
#VOLTAGE-TIME
import numpy as np
t=[i/10 for i in range(1,21)]
t
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title("Time vs Voltage")
plt.plot(t,v)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.grid()
plt.savefig('images/Time vs Voltage.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013

import pandas as pd
df=pd.read_csv('DataFiles/Income.csv')
years=list(range(2005,2014))
#df


incomeArkansas=df.values[3,2:]
plt.title('Income of Akansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of all states from 2005 to 2013

df
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())

    
plt.plot(years,incomeallstates)
plt.title('All States Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid()
plt.show()
```
